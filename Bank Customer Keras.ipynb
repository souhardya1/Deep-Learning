```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Raw Data

```python
df=pd.read_csv('Bank Customer Data.csv')
```

```python
df
```

## Preprocessing

```python
df.drop('RowNumber',axis=1,inplace=True)
```

```python
df
```

```python
df.drop('Surname',axis=1,inplace=True)
```

```python
df
```

```python
df.drop('CustomerId',axis=1,inplace=True)
```

```python
df
```

```python
Gender=pd.get_dummies(df['Gender'],drop_first=True) 
Geo=pd.get_dummies(df['Geography'],drop_first=True)
```

```python
df=pd.concat([df,Gender,Geo],axis=1)
```

```python
df
```

```python
df.drop(['Geography','Gender'],axis=1,inplace=True)
```

```python
df
```

## Input Output Split

```python
y=df['Exited']
X=df.drop('Exited',axis=1)
```

```python
y
```

```python
X
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

## Creating ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LeakyReLU,PReLU,ELU
```

#### Initialising the ANN

```python
classifier=Sequential()
```

#### Adding Input Layer and First Hidden Layer

```python
classifier.add(Dense(units=6, kernel_initializer='he_uniform',activation='relu',input_dim=11))
```

#### Adding Second Hidden Layer

```python
classifier.add(Dense(units=6, kernel_initializer='he_uniform',activation='relu'))
```

#### Adding Output Layer

```python
classifier.add(Dense(units=1, kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.summary()
```

#### Compiling ANN

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

#### Fitting the ANN to Train data

```python
model=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

#### Predicting on test data

```python
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```

#### Visualization of loss

```python
plt.plot(model.history['loss'])
plt.plot(model.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```
